### 2b. Timeframe : only take recent 3 years loan data : 2018 to 2020

```python
loans[['issue_d']].value_counts()
```

```python
def loan_years(x, y):
    
    k = []

    for i in range(x, y+1):
        m1 = 'Jan' + '-' + str(i)
        m2 = 'Feb' + '-' + str(i)
        m3 = 'Mar' + '-' + str(i)
        m4 = 'Apr' + '-' + str(i)
        m5 = 'May' + '-' + str(i)    
        m6 = 'Jun' + '-' + str(i) 
        m7 = 'Jul' + '-' + str(i) 
        m8 = 'Aug' + '-' + str(i)
        m9 = 'Sep' + '-' + str(i)
        m10 = 'Oct' + '-' + str(i)
        m11 = 'Nov' + '-' + str(i)
        m12 = 'Dec' + '-' + str(i)

        k.append(m1)
        k.append(m2)
        k.append(m3)
        k.append(m4)
        k.append(m5)
        k.append(m6)
        k.append(m7)
        k.append(m8)
        k.append(m9)
        k.append(m10)
        k.append(m11)
        k.append(m12)
    
    
    return k
```

```python
year_3 = loan_years(2018, 2020)[:-3]
```

## Advice

- The Lending Club dataset is a great one to work with.  As you mentioned, there are many observations, and I like your idea to **cut down to the most recent years**.  Feel free to start with a subset of the data (in terms of both observations and features) as you build your MVP and baseline your model.  Then you can scale up to more data after that.
- I imagine this dataset has class imbalance issues; that is, likely more people pay back their loans than default.  Be sure to check out the class imbalance strategies that Rita will teach you about tomorrow :slightly_smiling_face:


Have you given any thought to the type of metric you will be using to judge your model?  **The metric should be selected based on the business focus**.  For example, if we want to make sure to only give loans to people we know will pay them back, we may favor **recall** to “catch” all of the positive default class.  But we may also want to make sure not to say no to too many people so that we can still give out loans and make money - then **precision** might be more important.  Or perhaps it’s a combination of both, which may lead you to **F1 or some other F-beta**.  Just give it some thought.


- Finally, some of the features in this dataset were created by the Lending Club/some financial institution.  For example, **interest rate and grade**.  These are likely good predictors of who will pay back their loans, but they are also features that were likely created by financial professionals who are already judging the applicants.  That is to say, **you may not have access to features like these when a new person applies for a loan**.  Just give it some thought on how your model will be used and what type of information you would like your model to have access too.  You can also try **modeling with and without these features** to see how big of a difference they make.
